```python
white.columns = white.columns.str.replace(' ','.')
red['type'] = 1
white['type'] = 0

wines = pd.concat([red, white], ignore_index=True)
```

```python
wines.head()
```

```python
from sklearn.model_selection import train_test_split

X = wines.drop(columns = ['type'])
y = wines['type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 45)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

model.add(Dense(12, activation = 'relu', input_shape = (12,)))
model.add(Dense(9, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))




```

```python
from tensorflow.keras.losses import BinaryCrossentropy

model.compile(optimizer = 'adam', loss = BinaryCrossentropy())
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
wines
```

```python
model.fit(X_train ,y_train, epochs = 10)
```

```python
Prediction = model.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score
Prediction_binary = (Prediction >= 0.5).astype(int)


accuracy_score(y_test, Prediction_binary)*100
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Actual Values', alpha=0.5)
plt.scatter(range(len(Prediction)), Prediction, color='red', label='Predicted Values', alpha=0.5)
plt.xlabel('Sample Index')
plt.ylabel('Class')
plt.title('Predictions vs Actuals')
plt.legend()
plt.show()
```
